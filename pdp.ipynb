```python
import matplotlib.pyplot as plt
from sklearn.inspection import plot_partial_dependence
from sklearn.datasets import load_diabetes
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Load a sample dataset (you can replace this with your own dataset)
data = load_diabetes()
X = data.data
y = data.target

# Train a model (you can replace this with your own model)
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model.fit(X, y)

# Specify the feature(s) for which you want to create PDPs
# In this example, we'll create a PDP for the first feature (feature index 0)
features = 'number_inpatient'

# Create the PDP plot
fig, ax = plt.subplots(figsize=(8, 6))
plot_partial_dependence(model, X, features, ax=ax)

# Customize plot labels and title
plt.xlabel('Feature Values')
plt.ylabel('Partial Dependence')
plt.title('Partial Dependence Plot for Feature 0')

# Show the plot
plt.show()
```
